```python
import pandas as pd

# Load your dataset (replace 'your_dataset.csv' with the actual file path or URL)
df = pd.read_csv("C:\\Users\\Admin\\Desktop\\technolab intern\\Train.csv")

# Display the information about the DataFrame
print("Original DataFrame Info:")
print(df.info())
print("\n")

# Data Encoding Process
# 1. One-Hot Encoding for Categorical Variables
df_encoded = pd.get_dummies(df, columns=['Item_Fat_Content', 'Outlet_Location_Type', 'Outlet_Type'], drop_first=True)

# Display the DataFrame after encoding
print("DataFrame after One-Hot Encoding:")
print(df_encoded.head())
print("\n")

# Optionally, you can save the encoded DataFrame to a new CSV file
# df_encoded.to_csv('encoded_dataset.csv', index=False)
```

```python
from sklearn.preprocessing import LabelEncoder
label_columns = ['Item_Fat_Content', 'Outlet_Location_Type', 'Outlet_Type']
label_encoder = LabelEncoder()

for col in label_columns:
    df[col + '_LabelEncoded'] = label_encoder.fit_transform(df[col])

print("DataFrame after Label Encoding:")
print(df[['Item_Fat_Content', 'Item_Fat_Content_LabelEncoded',
          'Outlet_Location_Type', 'Outlet_Location_Type_LabelEncoded',
          'Outlet_Type', 'Outlet_Type_LabelEncoded']].head())
print("\n")

```
