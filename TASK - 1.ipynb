```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
cleaned_train_data = pd.read_csv("cleaned_train.csv")
cleaned_test_data = pd.read_csv("cleaned_test.csv")
```

```python
print("\nCleaned Test dataset columns:")
print(cleaned_test_data.columns)
```

```python
label_encoder = LabelEncoder()
```

```python

if 'Gender' in cleaned_train_data.columns and 'City' in cleaned_train_data.columns:

    cleaned_train_data['Gender_encoded'] = label_encoder.fit_transform(cleaned_train_data['Gender'])
    cleaned_train_data['City_encoded'] = label_encoder.fit_transform(cleaned_train_data['City'])
```

```python

if 'Gender' in cleaned_test_data.columns and 'City' in cleaned_test_data.columns:
    
    cleaned_test_data['Gender_encoded'] = label_encoder.transform(cleaned_test_data['Gender'])
    cleaned_test_data['City_encoded'] = label_encoder.transform(cleaned_test_data['City'])
```

```python
print("\nCleaned Train dataset after label encoding:")
print(cleaned_train_data.head())
```

```python
print("\nCleaned Test dataset after label encoding:")
print(cleaned_test_data.head())
```
